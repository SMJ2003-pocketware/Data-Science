**Importing required libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from imblearn.over_sampling import SMOTE
from sklearn.metrics import accuracy_score,classification_report
```

```python
creditcard_data=pd.read_csv("creditcard.csv")
```

**Handling class imbalance and splitting the dataset**

```python
X=creditcard_data.drop('Class', axis=1)
y=creditcard_data['Class']
```

```python
# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
#Applying SMOTE
smote = SMOTE(random_state=42)
X_train_res, y_train_res = smote.fit_resample(X_train, y_train)
```

```python
#Taking only a small part of the dataset due to its huge size
sampled_data=creditcard_data.sample(frac=0.1,random_state=42)
```

```python
#Handle the imbalance on the sampled_data
X_sample = sampled_data.drop('Class', axis=1)
y_sample = sampled_data['Class']
```

```python
# Splitting the sampled dataset
X_train_sample, X_test_sample, y_train_sample, y_test_sample = train_test_split(X_sample, y_sample, test_size=0.3, random_state=42)
```

```python
# Apply SMOTE on the sampled data
X_train_res_sample, y_train_res_sample = smote.fit_resample(X_train_sample, y_train_sample)
```

**Model building**

```python
model=RandomForestClassifier(random_state=42)
model.fit(X_train_res_sample,y_train_res_sample)
```

**Evaluating the model**

```python
#Accuracy Score
y_pred=model.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
#Classification Report
y_pred_sample=model.predict(X_test_sample)
classification_report(y_test_sample, y_pred_sample)
```
